```python
from sklearn.pipeline import make_pipeline
from sklearn.random_projection import GaussianRandomProjection
from lpproj import LocalityPreservingProjection
def nearestn():
    return KNeighborsClassifier(n_neighbors=5, p=2, algorithm='auto', n_jobs=-1)
def dummy():
    return DummyClassifier()
def nearestn_gaussianrandomp():
    return make_pipeline(
        GaussianRandomProjection(n_components=200),
        KNeighborsClassifier(n_neighbors=5, p=2, algorithm='auto', n_jobs=-1)
    )
def nearestn_lpp():
    return make_pipeline(
        LocalityPreservingProjection(n_components=200),
        KNeighborsClassifier(n_neighbors=5, p=2, algorithm='auto', n_jobs=-1)
    )
def nearestn_imagenet():
    
    class F(object):
        def __init__(self):
            self.knn = KNeighborsClassifier(n_neighbors=5, p=2, algorithm='auto', n_jobs=-1)     
        def fit(self, X, y):
            self.knn.fit(self.transform(X), y)
            
        def transform(self, X):
            shape = (X.shape[0],) + data.img_dim
            X = X.reshape(shape)
            X = X.transpose((0, 2, 3, 1))
            f = extractor.transform(X)
            return f
        
        def predict(self, X):
            return self.knn.predict(self.transform(X))
        
        def predict_proba(self, X):
            return self.knn.predict_proba(self.transform(X))

    return F()
```

```python
candidates = [dummy, nearestn, nearestn_gaussianrandomp, nearestn_imagenet]
for c in candidates:
    print(c.__name__)
    skf = StratifiedKFold(y, n_folds=5, shuffle=True, random_state=2)
    accs = []
    for train, test in skf:
        clf = c()
        clf.fit(X[train], y[train])
        acc = ((clf.predict(X[test]) == y[test]).mean())
        accs.append(acc)
        print(acc)
    print(np.mean(accs), np.std(accs))
```
