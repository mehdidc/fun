```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

plt.ion()

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = X**2 - Y**2
Z = R

    
surf = ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis,
                         antialiased=False,
                         linestyle='dotted',
                         zorder=2)
#ax.set_zlim(-, 100)
#ax.set_xlim(-100, 100)
#ax.set_ylim(-100, 100)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

#fig.colorbar(surf, shrink=0.5, aspect=5)

x, y = np.random.uniform(-4, 4), np.random.uniform(-4, 4)
alpha = 0.01
all_x, all_y, all_z = [], [], []
for i in range(100):
    dx = 2 * x
    x -= alpha * dx
    dy = -2 * y
    y -= alpha * dy
    z = x**2 - y**2
    all_x.append(x)
    all_y.append(y)
    all_z.append(z)
    #ax.scatter3D([x], [y], [z], c="red", zorder=0)
ax.scatter3D(all_x, all_y, all_z, c="red", zorder=0)

plt.show()
```
