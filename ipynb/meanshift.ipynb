```python
%matplotlib inline
import os
os.environ["DATA_PATH"] = "/root/work/data"
from lasagnekit.datasets.mnist import MNIST
import matplotlib.pyplot as plt
```

```python
data = MNIST()
data.load()
```

```python
X = data.X[0:10000]
```

```python
from sklearn.cluster import MeanShift
```

```python
clus = MeanShift(bandwidth=0.5)
```

```python
clus.fit(X)
```

```python
clus.cluster_centers_.shape
```

```python
for i in range(len(clus.cluster_centers_)):
    c = clus.cluster_centers_[i].reshape((28, 28))
    plt.imshow(c, cmap='gray', interpolation='none')
    plt.show()
```
