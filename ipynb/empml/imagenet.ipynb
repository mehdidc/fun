```python
%matplotlib inline
from lasagnekit.datasets.imagenet import ImageNet
import numpy as np
import matplotlib.pyplot as plt
from caffezoo.googlenet import GoogleNet
```

```python
classes = open("/root/work/data/imagenet/classnames_10000.txt").readlines()
```

```python
model = GoogleNet()
```

```python
model._load()
```

```python
data = ImageNet(nb=100)
data.load()
```

```python
idx = np.random.randint(0, data.X.shape[0])
plt.imshow(data.X[idx])
print(data.y[idx])
```

```python
prob = model.transform(data.X[idx:idx+1]*255)[0]
c = np.argsort(-prob)
for a in c[0:10]:
    print(classes[a])
```
